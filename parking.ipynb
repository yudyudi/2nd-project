## 2.1 임대보증금, 임대료 데이터타입 변환

```python
trains.info()
```

```python
trains[trains['임대보증금']=='-'] = trains[trains['임대보증금']=='-'].replace('-',np.NAN)
tests[tests['임대보증금']=='-'] = tests[tests['임대보증금']=='-'].replace('-',np.NAN)

trains[trains['임대료']=='-'] = trains[trains['임대료']=='-'].replace('-',np.NAN)
tests[tests['임대료']=='-'] = tests[tests['임대료']=='-'].replace('-',np.NAN)
```

```python
trains['임대보증금'] = trains['임대보증금'].astype(float)
tests['임대보증금'] = tests['임대보증금'].astype(float)

trains['임대료'] = trains['임대료'].astype(float)
tests['임대료'] = tests['임대료'].astype(float)
```

## 2.2 임대보증금, 임대료 null값 mean값으로 대체

```python
trains[trains.임대보증금.notnull()].자격유형.value_counts()
```

```python
t_mean = trains['임대보증금'].mean()
t_mean = np.round(t_mean)

te_mean = tests['임대보증금'].mean()
te_mean = np.round(te_mean)

t_mean, te_mean
```

```python
trains['임대보증금'] = trains['임대보증금'].fillna(t_mean)
tests['임대보증금'] = tests['임대보증금'].fillna(te_mean)
```

```python
tr_mean = trains['임대료'] .mean()
tr_mean = np.round(tr_mean)

tes_mean = tests['임대료'] .mean()
tes_mean = np.round(tes_mean)

tr_mean , tes_mean
```

```python
trains['임대료'] = trains['임대료'].fillna(tr_mean)
tests['임대료'] = tests['임대료'].fillna(tes_mean)
```

```python
# 임대보증금, 임대료 null 값이 없는것 확인
trains.isnull().sum()
```

```python
# 임대보증금, 임대료 null 값이 없는것 확인
tests.isnull().sum()
```

## 2.3 버스,지하철 null값 0으로 대체

```python
trains['도보 10분거리 내 지하철역 수(환승노선 수 반영)'] = trains['도보 10분거리 내 지하철역 수(환승노선 수 반영)'].fillna(0)
tests['도보 10분거리 내 지하철역 수(환승노선 수 반영)'] = tests['도보 10분거리 내 지하철역 수(환승노선 수 반영)'].fillna(0)

trains['도보 10분거리 내 버스정류장 수'] = trains['도보 10분거리 내 버스정류장 수'].fillna(0)
tests['도보 10분거리 내 버스정류장 수'] = tests['도보 10분거리 내 버스정류장 수'].fillna(0)
```
